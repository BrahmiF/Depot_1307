## Exercice 1 (3 pts)

Écrire une fonction qui reçoit un nombre `x` et retourne :
- `'Positif'` si `x > 0`
- `'Négatif'` si `x < 0`
- `'Zéro'` sinon



```python
def signe(x):
    """
    Retourne 'Positif', 'Négatif' ou 'Zéro' selon la valeur de x.
    """
    if x > 0:
        return 'Positif'
    elif x < 0:
        return 'Négatif'
    else:
        return 'Zéro'
```

## Exercice 2 (3 pts)

Écrire une fonction qui reçoit un âge (entier) et retourne :
- `'Enfant'` si âge < 12
- `'Adolescent'` si 12 ≤ âge ≤ 17
- `'Adulte'` si âge ≥ 18



```python

def categorie_age(age):
    """
    Retourne la catégorie d'âge selon l'âge donné.
    """
    if age < 12:
        return 'Enfant'
    elif 12 <= age <= 17:
        return 'Adolescent'
    else:
        return 'Adulte'
```

## Exercice 3 (3 pts)

Écrire une fonction qui reçoit une note sur 100 et retourne :
- `'Échec'` si note < 60
- `'Passable'` si 60–69
- `'Bien'` si 70–79
- `'Très bien'` si ≥ 80

## Exercice 4 (2 pts)

Écrire une fonction qui reçoit deux nombres `a` et `b` et retourne :
- le plus grand des deux
- ou la chaîne `'Égaux'` s’ils sont égaux



```python

def plus_grand_ou_egaux(a, b):
    """
    Retourne le plus grand des deux nombres ou 'Égaux' s'ils sont égaux.
    """
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return 'Égaux'
```

## Exercice 5 (4 pts)

Écrire une fonction qui reçoit un entier `n` et retourne une **liste de deux chaînes** :
1) `'Pair'` ou `'Impair'`
2) `'Multiple de 3'` ou `'Pas multiple de 3'`



```python
def pair_et_multiple(n):
    """
    Retourne une liste avec deux éléments : 
    - Pair/Impair
    - Multiple de 3 / Pas multiple de 3
    """
    resultat = []
    
    # Pair ou impair
    if n % 2 == 0:
        resultat.append('Pair')
    else:
        resultat.append('Impair')
    
    # Multiple de 3 ou non
    if n % 3 == 0:
        resultat.append('Multiple de 3')
    else:
        resultat.append('Pas multiple de 3')
    
    return resultat
```

## Exercice 7 (2 pts)

Écrire une fonction qui retourne la liste des nombres pairs entre 1 et 50 (inclus) avec une boucle `for`.



```python
def nombres_pairs_jusqu_a_50():
    """
    Retourne la liste des nombres pairs entre 1 et 50 inclus.
    """
    pairs = []
    for i in range(1, 51):
        if i % 2 == 0:
            pairs.append(i)
    return pairs
```

## Exercice 8 (2 pts)

Écrire un programme qui reçoit `n` et affiche la table de multiplication de `n` (de 1×n à 10×n).



```python
def table_multiplication(n):
    """
    Affiche la table de multiplication de n de 1 à 10.
    """
    for i in range(1, 11):
        print(f"{i} × {n} = {i * n}")
```

## Exercice 9 (3 pts)

Écrire un programme qui calcule et retourne la somme des nombres de 1 à 100 (inclus) en utilisant une boucle `for`.



```python
def somme_1_a_100():
    """
    Calcule et retourne la somme des entiers de 1 à 100.
    """
    somme = 0
    for i in range(1, 101):
        somme += i
    return somme
```

## Exercice 10 (2 pts)

Écrire une fonction qui reçoit une chaîne `mot` et retourne la liste de ses caractères (un caractère par élément) en utilisant une boucle `for`.


```python
def liste_caracteres(mot):
    """
    Retourne la liste des caractères d'une chaîne donnée.
    """
    caracteres = []
    for lettre in mot:
        caracteres.append(lettre)
    return caracteres
```

## Exercice 11 (2 pts)

Écrire un programme qui retourne la liste des nombres de 10 à 1 (inclus) en utilisant une boucle `while`.



```python
# Boucle while pour générer la liste de 10 à 1
nombres_descendants = []
i = 10
while i >= 1:
    nombres_descendants.append(i)
    i -= 1
print(nombres_descendants)
```

## Exercice 12 (2 pts)

Écrire une fonction qui reçoit une liste de nombres saisis successivement et retourne la somme jusqu’à rencontrer un 0.

Exemple : `[5, 2, -1, 0, 100]` -> somme = `6`.

Consigne : utiliser une boucle `while`.




```python
def somme_jusqu_a_zero(liste):
    """
    Calcule la somme des éléments d'une liste jusqu'à rencontrer un 0.
    """
    somme = 0
    i = 0
    while i < len(liste) and liste[i] != 0:
        somme += liste[i]
        i += 1
    return somme
```
